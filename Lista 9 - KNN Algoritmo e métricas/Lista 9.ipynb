```python
# using shuffle and random_random state in a stratified
cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)

pipe_Log = make_pipeline(StandardScaler(), KNeighborsClassifier())

param_weight = ['uniform','distance']
param_neighbors = [1,3,5,7,9]
param_leaf = [5,10,20,30,40,50,60,80,100,150,200,500,800,1000]

param_grid = [{'kneighborsclassifier__n_neighbors': param_neighbors,
               'kneighborsclassifier__weights': param_weight,
               'kneighborsclassifier__leaf_size': param_leaf,
               'kneighborsclassifier__algorithm': ['auto']},
              {'kneighborsclassifier__n_neighbors': param_neighbors,
               'kneighborsclassifier__weights': param_weight,
               'kneighborsclassifier__leaf_size': param_leaf,
               'kneighborsclassifier__algorithm': ['ball_tree']},
              {'kneighborsclassifier__n_neighbors': param_neighbors,
               'kneighborsclassifier__weights': param_weight,
               'kneighborsclassifier__leaf_size': param_leaf,
               'kneighborsclassifier__algorithm': ['kd_tree']},
              {'kneighborsclassifier__n_neighbors': param_neighbors,
               'kneighborsclassifier__weights': param_weight,
               'kneighborsclassifier__leaf_size': param_leaf,
               'kneighborsclassifier__algorithm': ['brute']}]

gs = GridSearchCV(estimator=pipe_Log,
                  param_grid=param_grid,
                  scoring='accuracy',
                  cv=cv,
                  n_jobs=-1,
                  return_train_score=True)

gs = gs.fit(X_train, y_train)
print('Best score: ', gs.best_score_)
print('Best params: ', gs.best_params_)
```

## Euclidian Distance

Dados dois valores, por exemplo, $p = (p_1,...,p_n)$ e $q = (q_1,...,q_n)$ a distância euclideana é dada por: $ \sqrt{(p_1-q_1)^2 + ... + (p_n-q_n)^2}$

```python
def euclidean_distance(X, row):
    X_ = (X - row) ** 2
    return (np.sum(X_, axis=1))**0.5 # <--- sqrt
```

## Manhattam Distance

Dados dois valores, por exemplo, $p = (p_1,...,p_n)$ e $q = (q_1,...,q_n)$ a distância de manhattan é dada por: $ |(p_1-q_1)| + ... + |(p_n-q_n)|$

```python
def manhattan_distance(X, row):
    X_ = np.abs(X - row)
    return np.sum(X_, axis=1)
```

##  Minkowski Distance

Dados dois valores, por exemplo, $p = (p_1,...,p_n)$ e $q = (q_1,...,q_n)$ a distância euclideana é dada por: $ (|(p_1-q_1)|^p + ... + |(p_n-q_n)|^p) ^ \frac {1}{p} $

```python
def minkowski_distance(X, row, p):
    X_ = (X - row) ** p
    return (np.sum(X_, axis=1))**(1/p) 
```

## Chebyshev Distance

Dados dois valores, por exemplo, $p = (p_1,...,p_n)$ e $q = (q_1,...,q_n)$ a distância euclideana é dada por: $max (|(p_1-q_1)| + ... + max|(p_n-q_n)|)$

```python
def chebyshev_distance(X, row):
    X_ = (X - row)
    return np.max(X_, axis=1)
```

# Implementation of KNN to classification and regression with brute force method


```python
def obter_idx_kNN(X, row, k):
    dist_euc = euclidean_distance(X, row)
    idx_sort = np.argsort(dist_euc)
    return idx_sort[0:k]
```

```python
idx_kNN = obter_idx_kNN(X, row, k=3)
idx_kNN

df['type'][idx_kNN] # setosa type
```

### Classification KNN

```python
# classification
def classifica(X, y, row, k):
    idx_kNN = obter_idx_kNN(X, row, k=k)
    count = np.bincount(y[idx_kNN])
    return np.argmax(count)
```

```python
def regressao(X, y, row, k):
    idx_kNN = obter_idx_kNN(X, row, k=k)
    return np.mean(y[idx_kNN])
```

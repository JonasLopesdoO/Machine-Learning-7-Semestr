```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('dataset.csv', header=None)
```

```python
df.head()
```

```python
plt.scatter(df[0], df[1], s=10)
```

```python
cluster = KMeans(n_clusters=3, random_state=42)
```

```python
result = cluster.fit(df)
```

```python
# Coisas importantes, se aparecer um pontinho novo, você precisa saber os centroides
print('Inertia:', cluster.inertia_, '\n')
print('Labels:\n', cluster.labels_, '\n')
print('Centroides:\n', cluster.cluster_centers_)
```

```python
centroides = cluster.cluster_centers_
```

```python
# Plotando gráfico juntamente com os centroides
plt.scatter(df[0], df[1], s=10, c=cluster.labels_)
plt.scatter(centroides[:, 0], centroides[:, 1], marker='^', c='r', s=100)
```

```python
inertias = []

for k in range (1, 21):
    cluster = KMeans(n_clusters=k, random_state=42)
    cluster.fit(df)
    inertias.append(cluster.inertia_)
```

```python
x = np.arange(1, 21)
```

```python
# Plotando a inércia (cotovelo), percebemos que 3 é o número ideal para k.
plt.plot(x, inertias, '.-')
plt.xticks(x);
```

## ---------------------------------------------------------------------------------------------------------------------------------

## Preprocessing of data

### Transforming labels to numeric values

```python
# df.isnull() é a mesma coisa da linha a baixo
df_final.isna()
```

```python
# pega a quantidade de valores nulos para cada coluna
df_final.isna().sum()
```

```python
df_final = df_final.dropna(how='all') # elimina apenas quando todos os valores de uma linha são NaN
# Elimina colunas com 3 ou mais valores NaN
df_final = df_final.dropna(thresh=3)
```

```python
df_final
```

```python
df_final['LABEL'].value_counts()
```

```python
# create dictionary where labels_num and sex_num are a map to categoric values
labels_num = { v: k for k, v in enumerate(df_final['LABEL'].unique()) }
sex_num = { v: k for k, v in enumerate(df_final['SEX'].unique()) }

df_final['LABEL'] = df_final['LABEL'].map(labels_num)
df_final['SEX'] = df_final['SEX'].map(sex_num)

df_final[7831:7861]
```

```python
df_final['LABEL'].unique()
```

```python
# using Label Encoder to set non ordered features
le = LabelEncoder()
df_final['WORKCLASS'] = le.fit_transform(df_final['WORKCLASS'])
df_final['EDUCATION'] = le.fit_transform(df_final['EDUCATION'])
df_final['MERITAL_STATUS'] = le.fit_transform(df_final['MERITAL_STATUS'])
df_final['OCCUPATION'] = le.fit_transform(df_final['OCCUPATION'])
df_final['RELATIONSHIP'] = le.fit_transform(df_final['RELATIONSHIP'])
df_final['RACE'] = le.fit_transform(df_final['RACE'])
df_final['NATIVE_COUNTRY'] = le.fit_transform(df_final['NATIVE_COUNTRY'])
df_final
```

```python
# axis 0 about columns
prep = Imputer(missing_values='?', strategy='mean', axis=0)
prep = prep.fit(df_final)
df_final = prep.transform(df_final)
df_final

# por algum motivo ele não reconhece o '?' como um valor faltante, pela falta de tempo  não vai dar para implementar agora :/
```

### Setting dataset for trainning and test

```python
dados = resample.Utils.split_stratified_train_test(df_final, 0.7, 42)
X_train = dados[0]
y_train = dados[1]
X_test = dados[2]
y_test = dados[3]

# defining train and test for X normalized
X_train_norm = transform.Normalize.fit_transform(X_train)
X_test_norm = transform.Normalize.fit_transform(X_test)


# defining train and test for X standardized
X_train_std = transform.Standardize.fit_transform(X_train)
X_test_std = transform.Standardize.fit_transform(X_test)
```

# Creating train and validation of models using Accuracy metric

### -------------------------------------------------------------------------------------------------------------------------------------------------------------

### Used classifiers:

- Linear Regression - Scikit
- Stochastic Gradient Descent Regressor (SGDRegressor) do Scikit-Learn

### -------------------------------------------------------------------------------------------------------------------------------------------------------------

### Test

```python
'''Setting model variables'''
model_Linear_Regression = linear_model.LinearRegression()
model_SGDRegressor = linear_model.SGDRegressor(max_iter=1000, tol=1e-3)

'''Setting model variables normalized'''
model_Linear_Regression_norm = linear_model.LinearRegression()
model_SGDRegressor_norm = linear_model.SGDRegressor(max_iter=1000, tol=1e-3)

'''Setting model variables standardized'''
model_Linear_Regression_std = linear_model.LinearRegression()
model_SGDRegressor_std = linear_model.SGDRegressor(max_iter=1000, tol=1e-3)
```

### -------------------------------------------------------------------------------------------------------------------------------------------------------------

### Trainning models

```python
'''Trainning models normaly'''
model_Linear_Regression.fit(X_train, y_train)
model_SGDRegressor.fit(X_train, y_train)

'''Trainning models with normatization'''
model_Linear_Regression_norm.fit(X_train_norm, y_train)
model_SGDRegressor_norm.fit(X_train_norm, y_train)

'''Trainning models with standardization'''
model_Linear_Regression_std.fit(X_train_std, y_train)
model_SGDRegressor_std.fit(X_train_std, y_train);
```

### -------------------------------------------------------------------------------------------------------------------------------------------------------------

## Make predictions

```python
'''Make predictions with X_test normaly'''
pred_Linear_Regression = model_Linear_Regression.predict(X_test)
pred_SGDRegression = model_SGDRegressor.predict(X_test)

'''Make predictions with X_test normatized'''
pred_Linear_Regression_norm = model_Linear_Regression.predict(X_test_norm)
pred_SGDRegression_norm = model_SGDRegressor.predict(X_test_norm)

'''Make predictions with X_test standardized'''
pred_Linear_Regression_std = model_Linear_Regression.predict(X_test_std)
pred_SGDRegression_std = model_SGDRegressor.predict(X_test_std)
```
